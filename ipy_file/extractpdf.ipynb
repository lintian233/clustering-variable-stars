```python
import camelot
import seaborn as sns  #笔者jupyter背景较暗,此库只用来明亮背景
```

```python
sns.set()

def extract_table(filepath,pages,table_area=['140,700,500,240']):
    table = camelot.read_pdf(filepath,
                                flavor="stream",
                                pages=pages,
                                table_areas = table_area)
    return table[0]

table = extract_table(r"../datasets/full_text.pdf",'4')
plt = camelot.plot(table,kind="textedge")
plt.show()


extract_table(r"../datasets/full_text.pdf",'4',['140,710,550,240']).df
```
